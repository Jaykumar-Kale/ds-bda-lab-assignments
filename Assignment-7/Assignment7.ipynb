## OBJECTIVES
1. To understand data visualization techniques.
2. To analyze patterns using graphical representation.
3. To apply visualization on Heart and Air Quality datasets.

Data visualization helps in:
- Identifying patterns
- Detecting outliers
- Understanding distribution
- Finding relationships between variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

## Step 1: Load Datasets
We load the Heart dataset and Air Quality dataset.

```python
heart_df = pd.read_csv("heart.csv")
air_df = pd.read_csv("air_quality.csv")

heart_df.head()
```

## 1️⃣ Pie Chart

A pie chart represents proportional distribution of categories.

Here we visualize the percentage of patients with and without heart disease.

```python
heart_df['target'].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=['skyblue','lightcoral']
)

plt.title("Heart Disease Distribution")
plt.ylabel("")
plt.show()
```

## 2️⃣ Bar Chart

Bar chart compares categorical data.

We visualize chest pain type distribution.

```python
sns.countplot(x='cp', data=heart_df)
plt.title("Chest Pain Type Distribution")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.show()
```

## 3️⃣ Histogram

Histogram shows frequency distribution of continuous data.

We plot age distribution.

```python
plt.hist(heart_df['age'], bins=5, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## 4️⃣ Box Plot

Boxplot shows:
- Minimum
- Q1
- Median
- Q3
- Maximum
- Outliers

We compare cholesterol levels based on heart disease.

```python
sns.boxplot(x='target', y='chol', data=heart_df)
plt.title("Cholesterol vs Heart Disease")
plt.show()
```

## 5️⃣ Line Chart

Line chart shows trend over ordered values.

We visualize SO2 levels across cities.

```python
plt.plot(air_df['so2'], marker='o')
plt.title("SO2 Levels Trend")
plt.xlabel("City Index")
plt.ylabel("SO2 Level")
plt.show()
```

## 6️⃣ Scatter Plot

Scatter plot shows relationship between two continuous variables.

We analyze Age vs Cholesterol.

```python
plt.scatter(heart_df['age'], heart_df['chol'])
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

## 7️⃣ Scatter Plot Matrix

Pairplot shows relationship between multiple continuous variables.

```python
sns.pairplot(heart_df[['age','chol','thalach','oldpeak']])
plt.show()
```

## 8️⃣ Correlation Heatmap

Heatmap shows correlation between numerical variables.

```python
plt.figure(figsize=(8,6))
sns.heatmap(heart_df.corr(numeric_only=True), 
            annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```
